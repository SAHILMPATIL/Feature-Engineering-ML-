```python
#imputation missing technique
sns.histplot(df['age'],kde=True)
```

```python
df['Age_mean']=df['age'].fillna(df['age'].mean())
```

```python
df[['Age_mean','age']]
```

mean imputation works well when we have normally distributed data


Next technique is Median Value Imputation (this is used when we have outliers in the dataset)

```python
df['age_median']=df['age'].fillna(df['age'].median())
```

```python
df[['age_median','Age_mean','age']]
```

NExt is MOde imputation technique for categorical values

```python
df[df['embarked'].isnull()]#embarked is an example of MCAR
```

```python
df['embarked'].unique()
```

trying to remove nan values

```python
mode_value=df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_value)
```

```python
df[['embarked_mode','embarked']]#to display the columns to verify that we have replaced the Nan values with mode
```

```python
df['embarked_mode'].isnull().sum()#checks for the null values in Embarked_mode  column
```

here np.int64(0) refers that there are no Nan (missing) values

```python
df['embarked'].isnull().sum()#checks for the null values in Embarked column
```

here np.int64(2) refers that there are 2 Nan (missing) values
